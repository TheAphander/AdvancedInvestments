```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
dataset = pd.read_excel("Assignment 1 Data Advanced Investment.xlsx")
```

### Regression 1

```python
wml_returns_decimals = (dataset.values[:,14] - dataset.values[:,5])
I_u = (dataset.values[:,1]> dataset.values[:,5]).astype(int)
I_b = np.zeros_like(wml_returns_decimals)
market_excess_returns_decimals = dataset.values[:, 1]

window_size = 24

# Initialize I_b array with zeros (same size as the dataset)
I_b = np.zeros(len(market_excess_returns_decimals), dtype=int)

# Loop through each value in the dataset
for i in range(len(market_excess_returns_decimals)):
    # Determine the start index for the rolling window
    start_idx = max(0, i - window_size)
    
    # Calculate the cumulative return over the rolling window
    rolling_cumulative_return = np.prod(1 + (market_excess_returns_decimals[start_idx:i]/100)) - 1
    
    # Set I_b to 1 if the cumulative return is negative, otherwise 0
    I_b[i] = 1 if rolling_cumulative_return < 0 else 0
wml_returns_decimals = np.array([float(x) for x in wml_returns_decimals])
market_excess_returns_decimals = np.array([float(x) for x in market_excess_returns_decimals])
X = sm.add_constant(market_excess_returns_decimals)  # Adds a column of ones for alpha_0

# Perform OLS regression
model = sm.OLS(wml_returns_decimals, X).fit()

# Print the summary of the regression
print(model.summary())
```

### Regression 2

```python
interaction_alpha = I_b  # I_b term for alpha_B
interaction_beta = I_b * market_excess_returns_decimals  # I_b * R_m term for beta_B

# Combine predictors
X = np.column_stack([
    np.ones(len(wml_returns_decimals)),  # Constant (alpha_0)
    interaction_alpha,                   # I_b (alpha_B)
    market_excess_returns_decimals,      # R_m (beta_0)
    interaction_beta                     # I_b * R_m (beta_B)
])

# Label the predictors for clarity in the summary
predictor_labels = ['alpha_0', 'alpha_B', 'beta_0', 'beta_B']

model = sm.OLS(wml_returns_decimals, X).fit()

# Print the summary with labeled predictors
model_summary = model.summary(xname=predictor_labels)
print(model_summary)
```

### Regression 3

```python
# Create interaction terms
interaction_alpha = I_b  # I_{B,t-1} for alpha_B
interaction_beta = I_b * market_excess_returns_decimals  # I_{B,t-1} * R_m for beta_B
interaction_beta_U = I_u * interaction_beta  # I_{B,t-1} * I_{U,t} * R_m for beta_{B,U}

# Combine predictors
X = np.column_stack([
    np.ones(len(wml_returns_decimals)),  # Constant (alpha_0)
    interaction_alpha,                   # I_{B,t-1} (alpha_B)
    market_excess_returns_decimals,      # R_m (beta_0)
    interaction_beta,                    # I_{B,t-1} * R_m (beta_B)
    interaction_beta_U                   # I_{B,t-1} * I_{U,t} * R_m (beta_{B,U})
])

# Label the predictors for clarity in the summary
predictor_labels = ['alpha_0', 'alpha_B', 'beta_0', 'beta_B', 'beta_B,U']

# Perform the regression
model = sm.OLS(wml_returns_decimals, X).fit()

# Print the summary with labeled predictors
print(model.summary(xname=predictor_labels))

```
